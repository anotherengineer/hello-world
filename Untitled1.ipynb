```python
# Jayachandra Shivapla
# Question 2 a

import numpy as np
import matplotlib.pyplot as plt

# parameters
E = 10 
r = 0.1 
sigma = 0.4 
T = 1/3
Nplus = 10 
Nminus = 0 
M = 100
dt = T/M  
omega = 1




def PSOR_solver_American(u,b,g,Nminus, Nplus , omega, eps):
    
    
    u[Nminus] = g[Nminus]
    u[Nplus] = g[Nplus]
    
    while True:
        error = 0.0
        
        for n in range(Nminus+1 , Nplus):
            aj=0.25*pow(sigma,2)*pow(n,2)*dt + 0.25*r*n*dt
            bj=0.5*pow(sigma,2)*pow(n,2)*dt + 0.5*r*dt
            cj=0.25*pow(sigma,2)*pow(n,2)*dt - 0.25*r*n*dt
            y=(1/(1+bj))*(b[n]+aj*u[n+1]+cj*u[n-1])
            y = max(g[n] , u[n]+omega*(y - u[n]))
            error += (u[n] - y)* (u[n] - y)
            u[n] = y
                 
        
        if ( error > eps):
                 break
    return u;
                 
                 
def American( dt , dx , Nminus , Nplus , M , omega , eps):
    
    
    
    x = np.zeros( Nplus - Nminus + 1)
    u = np.zeros ( Nplus - Nminus + 1)
    g = np.zeros ( Nplus - Nminus + 1)
    b = np.zeros ( Nplus - Nminus + 1)
    for n in range(Nminus,Nplus+1):
        x[n]  = n*dx
        u[n] = payoffPut(x[n])
        
    for m in range(1,M+1):
        tau = m*dt
        for n in range(Nminus+1, Nplus):
            g[n] = payoffPut(x[n])
            aj=0.25*pow(sigma,2)*pow(n,2)*dt + 0.25*r*n*dt
            bj=0.5*pow(sigma,2)*pow(n,2)*dt + 0.5*r*dt
            cj=0.25*pow(sigma,2)*pow(n,2)*dt - 0.25*r*n*dt
            b[n]= u[n]*(1-bj)+u[n+1]*aj+u[n-1]*cj     
            
        g[Nminus] = payoffPut(x[Nminus])
        g[Nplus] = payoffPut(x[Nplus])
        u[Nminus] = g[Nminus]
        u[Nplus] = g[Nplus]
        
        u = PSOR_solver(u,b,g,Nminus,Nplus,omega,eps)
    return u

def payoffCall( s_T , E = E):
    return max([s_T - E,0])

def payoffPut( s_T , E = E):
    return max([E - s_T , 0])


AmericanPutValues = American(dt , 2 , Nminus , Nplus , M , omega , pow(10,-8))

print( AmericanPutValues)
              
```
